```python
import re


def extract_year(title):
    match = re.search(r"\((\d{4})\)$", title)
    if match:
        return int(match.group(1))
    return None 

movie_info = {}
with open(r"C:\Users\punam\OneDrive\Desktop\ml-1m\movies.dat") as movies:
    for line in movies:
        columns = line.strip().split("::")
        movie_id = int(columns[0])
        title = columns[1]
        genres = columns[2].split("|")
        release_year = extract_year(title)  
        

        movie_info[movie_id] = {'title': title, 'genres': genres, 'release_year': release_year}

min_few_ratings = 10

few_rated_movies = {movie_id: count for movie_id, count in movie_rating_count.items() if count < min_few_ratings}

popular_movies = {movie_id: movie_info[movie_id] for movie_id in movie_ratings if movie_rating_count[movie_id] >= min_few_ratings}
few_rated_movies_info = {movie_id: movie_info[movie_id] for movie_id in few_rated_movies}

from collections import Counter

def analyze_characteristics(movies):
    release_years = [movie['release_year'] for movie in movies.values()]
    genres = [genre for movie in movies.values() for genre in movie['genres']]
    
    release_year_count = Counter(release_years)
    
    genre_count = Counter(genres)
    
    return release_year_count, genre_count

popular_release_years, popular_genres = analyze_characteristics(popular_movies)

few_rated_release_years, few_rated_genres = analyze_characteristics(few_rated_movies_info)

print("\nRelease Year Distribution of Popular Movies:")
for year, count in sorted(popular_release_years.items()):
    print(f"{year}: {count} movies")

print("\nRelease Year Distribution of Few-Rated Movies:")
for year, count in sorted(few_rated_release_years.items()):
    print(f"{year}: {count} movies")

print("\nGenres of Popular Movies:")
for genre, count in popular_genres.items():
    print(f"{genre}: {count} movies")

print("\nGenres of Few-Rated Movies:")
for genre, count in few_rated_genres.items():
    print(f"{genre}: {count} movies")
```

```python
user_data = {}
with open(r"C:\Users\punam\OneDrive\Desktop\ml-1m\users.dat") as user_file:
    for line in user_file:
        columns = line.strip().split("::")
        user_id = int(columns[0])
        gender = columns[1]
        age = int(columns[2])
        occupation = columns[3]
        user_data[user_id] = {'gender': gender, 'age': age, 'occupation': occupation}

ratings = []
with open(r"C:\Users\punam\OneDrive\Desktop\ml-1m\ratings.dat") as rating_file:
    for line in rating_file:
        columns = line.strip().split("::")
        user_id = int(columns[0])
        movie_id = int(columns[1])
        rating = int(columns[2])
        ratings.append((user_id, movie_id, rating))

demographic_activity = {'gender': Counter(), 'age_group': Counter(), 'occupation': Counter()}

for user_id, movie_id, rating in ratings:
    if user_id in user_data:
        user_info = user_data[user_id]
        gender = user_info['gender']
        age = user_info['age']
        occupation = user_info['occupation']
        age_group = age // 10 * 10  # Group age into 10-year intervals

        demographic_activity['gender'][gender] += 1
        demographic_activity['age_group'][age_group] += 1
        demographic_activity['occupation'][occupation] += 1

print("\nRatings by Gender:")
for gender, count in demographic_activity['gender'].items():
    print(f"{gender}: {count} ratings")

print("\nRatings by Age Group:")
for age_group, count in sorted(demographic_activity['age_group'].items()):
    print(f"{age_group}s: {count} ratings")

print("\nRatings by Occupation:")
for occupation, count in demographic_activity['occupation'].items():
    print(f"{occupation}: {count} ratings")
```
